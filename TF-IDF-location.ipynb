## TF-IDF

### All SDI together ("Most frequent terms in month M as compared to months ~M for all tweets")

```python
from sklearn.feature_extraction import text
my_stop_words = text.ENGLISH_STOP_WORDS.union(["coronaviru", "covid", "im", "amp", "like", "ive"])
```

```python
# fit on months we're not looking at and transform on months separately

def tfidf_transform(month, ngram_range, df):
    vectorizer = TfidfVectorizer(ngram_range=ngram_range, min_df=0.01, max_df=0.9, stop_words=my_stop_words)
    vectorizer = vectorizer.fit([" ".join(x) for x in df[df['month'] != month]['stemmer']])
    feature_names = pd.Series(vectorizer.get_feature_names())
    
    tfidf = vectorizer.transform([" ".join(x) for x in df[df['month'] == month]['stemmer']])
    weights = pd.Series(np.array(tfidf.sum(axis=0)).flatten())
    top_words = list(weights.nlargest(n=100).index)
    
    return (feature_names.iloc[top_words].reset_index(drop=True))
```

```python
top_monograms_both = pd.DataFrame()
for month in ['January', 'February', 'March']:
    top_monograms_both[month] = tfidf_transform(month, ngram_range=(1,1), df=df)
```

```python
shared_top_words_both = list(set(top_monograms_both['January'][:50]).intersection(set(top_monograms_both['February'][:50]), set(top_monograms_both['March'][:50])))
```

```python
top_monograms_both
```

```python
top_monograms_both_filt = pd.DataFrame()
for month in ['January', 'February', 'March']:
    top_monograms_both_filt[month] = pd.Series([x for x in list(top_monograms_both[month]) if x not in shared_top_words_both])
```

## For each month, most frequent terms for low vs. high

```python
# for a given month, fit on both low and high SDI and transform on SDI DFs separately

def tfidf_low_v_high_transform(month, ngram_range, low, high):
    vectorizer = TfidfVectorizer(ngram_range=ngram_range, min_df=0.01, max_df=0.9, stop_words=my_stop_words)
    vectorizer = vectorizer.fit([" ".join(x) for x in low[low['month'] == month]['stemmer']])
    feature_names = pd.Series(vectorizer.get_feature_names())
    
    tfidf_low = vectorizer.transform([" ".join(x) for x in low[low['month'] == month]['stemmer']])
    weights_low = pd.Series(np.array(tfidf_low.sum(axis=0)).flatten())
    top_words_low = list(weights_low.nlargest(n=100).index)
    
    tfidf_high = vectorizer.transform([" ".join(x) for x in high[high['month'] == month]['stemmer']])
    weights_high = pd.Series(np.array(tfidf_high.sum(axis=0)).flatten())
    top_words_high = list(weights_high.nlargest(n=100).index)
    
    return (feature_names.iloc[top_words_low].reset_index(drop=True),
            feature_names.iloc[top_words_high].reset_index(drop=True))
```

```python
top_monograms = pd.DataFrame()
for month in ['January', 'February', 'March']:
    out = tfidf_low_v_high_transform(month, (1,1), low=low_sdi, high=high_sdi)
    shared_top_words = list(set(out[0][:50]).intersection(set(out[1][:50])))
    top_monograms[month + ' low_SDI'] = pd.Series([x for x in list(out[0]) if x not in shared_top_words])
    top_monograms[month + ' high_SDI'] = pd.Series([x for x in list(out[1]) if x not in shared_top_words])
```

```python
top_monograms[['March low_SDI', 'March high_SDI']]
```

```python
top_monograms = pd.DataFrame()
for month in ['January', 'February', 'March']:
    out = tfidf_low_v_high_transform(month, (1,1), low=df, high=hotspots_df)
    shared_top_words = list(set(out[0][:50]).intersection(set(out[1][:50])))
    top_monograms[month + ' all'] = pd.Series([x for x in list(out[0]) if x not in shared_top_words])
    top_monograms[month + ' hotspots'] = pd.Series([x for x in list(out[1]) if x not in shared_top_words])
```
